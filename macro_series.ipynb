```python
import numpy as np
import pandas as pd

import pickle

from sqlalchemy import create_engine
```

```python
query = """
    SELECT * FROM macro
    WHERE date > '2014-3-16'
    AND date < '2019-03-18'
"""

engine = create_engine('postgres://ubuntu:project-kojak@18.188.49.181:5432/roboadvisor')
```

```python
macro = pd.read_sql(query, engine)
```

```python
with open('codes.pickle', 'rb') as p:
    codes = pickle.load(p)
```

```python
type(macro['code'][0])
```

```python
len(codes)
```

```python
macro_short = macro[macro['code'].isin(codes)]
```

```python
macro_short
```

```python
macro_short.date = [pd.to_datetime((str(date).split())[0]) for date in macro_short.date]
```

```python
len(macro_short)
```

```python
macro['date']
```

```python
macro_pivot = macro_short.pivot(index='date', columns='code', values='value')
```

```python
macro_pivot
```

```python
macro_interp = macro_pivot.resample('D').interpolate(method='time')
```

```python
macro_interp = macro_interp.interpolate(method='time', limit_direction='backward')
```
