```python
import numpy as np
import pandas as pd

from fbprophet import Prophet
```

```python
etfs = pd.read_csv('etfs_pivot.csv')
etfs['date'] = pd.to_datetime(etfs['date'])
```

```python
first_year = pd.date_range(start = '2014-03-17', end = '2015-03-16').tolist()
second_year = pd.date_range(start = '2015-03-17', end = '2016-03-15').tolist()
third_year = pd.date_range(start = '2016-03-17', end = '2017-03-16').tolist()
fourth_year = pd.date_range(start = '2017-03-17', end = '2018-03-16').tolist()
fifth_year = pd.date_range(start = '2018-03-17', end = '2019-03-15').tolist()
```

```python
spy = etfs[['date','SPY']]
```

```python
spy_fourth = spy[spy['date'].isin(pd.date_range(first_year[0], fourth_year[-1]))]
spy_fourth.columns = ['ds', 'y']
```

```python
arima = Prophet()
arima.fit(spy_fourth)
future = arima.make_future_dataframe(periods = len(fifth_year))
forecast = arima.predict(future)
```

```python
forecast_fifth = forecast[forecast['ds'].isin(fifth_year)]['yhat']
```

```python
forecast_fifth
```

```python
spy_fifth = spy[spy['date'].isin(fifth_year)]
```

```python
spy_predict = pd.DataFrame(index = spy_fifth['date'])
spy_predict['actual'] = np.array(spy_fifth['SPY'])
spy_predict['forecast'] = np.array(forecast_fifth)
spy_predict['residual'] = spy_predict['actual'] - spy_predict['forecast']
```
