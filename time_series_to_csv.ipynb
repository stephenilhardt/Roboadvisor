```python
import numpy as np
import pandas as pd

from fbprophet import Prophet

from sqlalchemy import create_engine
```

```python
df = pd.read_csv('etfs_pivot.csv')
```

```python
df = df.set_index('date')
```

```python
etfs = df.columns
```

```python
len(etfs)
```

```python
etfs = list(etfs)
```

```python
predict_series = pd.date_range(start='2014-03-17', end='2018-03-16')
forecast_series = pd.date_range(start='2018-03-17', end='2019-03-15')
```

```python
time_series_etf_predictions = pd.DataFrame(index=forecast_series)
```

```python
for etf in etfs:
    etf_frame = df[etf]
    etf_frame = etf_frame.reset_index()
    etf_frame.columns = ['ds','y']
    
    arima_model = Prophet()
    arima_model.fit(etf_frame)
    
    future = arima_model.make_future_dataframe(periods=365)
    forecast = arima_model.predict(future)
    
    predictions = list(forecast[forecast['ds'].isin(forecast_series)]['yhat'])
    
    time_series_etf_predictions[etf] = predictions
```

```python
len(time_series_etf_predictions.columns)
```

```python
time_series_etf_predictions.to_csv('time_series_predictions.csv')
```

```python
time_series_etf_predictions = time_series_etf_predictions.reset_index()
```

```python
time_series_etf_predictions = time_series_etf_predictions.rename(columns={'index': 'date'})
```

```python
etf_time_series = time_series_etf_predictions.melt(id_vars='date').set_index('date')
```

```python
etf_time_series.to_csv('etf_arima.csv')
```

```python
etf_time_series = etf_time_series.rename(columns={'variable': 'symbol', 'value': 'arima_forecast'})
```

```python
etf_time_series.to_csv('etf_arima.csv')
```
