```python
import numpy as np
import pandas as pd

from fbprophet import Prophet
```

```python
first_year = pd.date_range(start = '2014-03-17', end = '2015-03-16').tolist()
second_year = pd.date_range(start = '2015-03-17', end = '2016-03-15').tolist()
third_year = pd.date_range(start = '2016-03-17', end = '2017-03-16').tolist()
fourth_year = pd.date_range(start = '2017-03-17', end = '2018-03-16').tolist()
fifth_year = pd.date_range(start = '2018-03-17', end = '2019-03-15').tolist()
```

```python
etfs = pd.read_csv('/home/ubuntu/csvs/etfs_pivot.csv')
```

```python
etfs = etfs.set_index('date')
```

```python
etfs.index = pd.to_datetime(etfs.index)
```

```python
etf_list = etfs.columns.tolist()
```

```python
len(etf_list)
```

```python
second_year_predictions = pd.DataFrame(index=second_year)
```

```python
third_year_predictions = pd.DataFrame(index=third_year)
fourth_year_predictions = pd.DataFrame(index=fourth_year)
fifth_year_predictions = pd.DataFrame(index=fifth_year)
```

```python
def make_fifth_year_predictions(etf):
    etf_df = pd.DataFrame(etfs[etf])
    etf_df = etf_df.reset_index()
    etf_df = etf_df[etf_df['date'].isin(fourth_year)]
    etf_df.columns = ['ds', 'y']

    model = Prophet()
    model.fit(etf_df)
    
    future = model.make_future_dataframe(periods=(len(fifth_year)+1))
    forecast = model.predict(future)
    
    fifth_year_etf_predictions = np.array(forecast[forecast['ds'].isin(fifth_year)]['yhat'])
    fifth_year_predictions[etf] = fifth_year_etf_predictions
    
    return -1
```

```python
for etf in etf_list:
    make_fifth_year_predictions(etf)
```

```python
fifth_year_predictions.to_csv('/home/ubuntu/csvs/fifth_year_arima.csv')
```
