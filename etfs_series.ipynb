```python
import numpy as np
import pandas as pd

from sqlalchemy import create_engine
```

```python
engine = create_engine('postgres://ubuntu:project-kojak@18.188.49.181:5432/roboadvisor')

etfs = pd.read_sql_table('etfs', engine)
```

```python
etfs = etfs.set_index('date')
```

```python
etfs
```

```python
etfs_pivot = etfs.pivot(columns='ticker', values='close')
```

```python
etfs_pivot = etfs_pivot.reset_index()
```

```python
print(etfs_pivot.date[2])
```

```python
etfs_pivot['date'] = [str(date) for date in etfs_pivot['index']]
```

```python
etfs_pivot.date = ["-".join(date.split('-')) for date in etfs_pivot.date]
```

```python
etfs_pivot['date'] = pd.to_datetime(etfs_pivot['date'], format='%Y-%m-%d')
```

```python
etfs_new = pd.DataFrame(etfs_pivot.values, columns=etfs_pivot.columns, index=etfs_pivot.date).drop('date',axis=1)
```

```python
etfs_reindex = etfs_new.reindex(pd.date_range(start='2014-03-17',end='2019-03-17', freq='D'), method='ffill')

etfs_reindex
```

```python
etfs_data = etfs_reindex.reset_index().rename({'index': 'date'}, axis=1).set_index('date')
```
