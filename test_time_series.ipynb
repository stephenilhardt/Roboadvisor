```python
import numpy as np
import pandas as pd

from fbprophet import Prophet
```

```python
etfs = pd.read_csv('etfs_pivot.csv')
```

```python
etfs
```

```python
etfs = etfs.reset_index()
etfs['date'] = pd.to_datetime(etfs['date'])

tickers = etfs.columns[1:]

etfs_short = etfs[etfs['date'] < '2018-3-15']
```

```python
tickers = tickers.tolist()
```

```python
etf_preds = pd.DataFrame(index = pd.date_range(start='2014-03-17', end='2019-03-15', freq='D'))
```

```python
etfs_short = etfs_short.set_index('date')
```

```python
for ticker in tickers:
    df = etfs_short[ticker]
    df = df.reset_index()
    df.columns = ['ds', 'y']
    model = Prophet()
    model.fit(df)
    future = model.make_future_dataframe(periods=366)
    
    preds = model.predict(future)
    
    etf_preds[ticker] = preds['yhat'].tolist()
```
