```python
import numpy as np
import pandas as pd

from sqlalchemy import create_engine
```

```python
engine = create_engine('postgres://ubuntu:project-kojak@localhost:5432/roboadvisor')
```

```python
query = '''SELECT prices.*, dividends.dividend FROM prices 
LEFT JOIN dividends ON prices.ticker = dividends.ticker AND prices.date = dividends.date;
'''

prices = pd.read_sql(query, engine, index_col = 'ticker')
```

```python
prices['date_split'] = [str(date).split() for date in prices['date']]
prices['date'] = [date[0] for date in prices['date_split']]
prices['date'] = pd.to_datetime(prices['date'])
```

```python
prices = prices.drop(['adjusted_close','date_split'], axis=1)
```

```python
prices
```

```python
prices = prices.fillna(0.0)
prices['split'] = 1.0
```

```python
tickers = prices.index
```

```python
tickers = sorted(list(tickers))
```

```python
prices = prices.sort_values(['ticker', 'date'])
```

```python
prices
```

```python
tickers_set = set(tickers)
tickers_list = list(tickers_set)
tickers = [str(ticker) for ticker in tickers_list]
```

```python
tickers = sorted(tickers)
```

```python
prices = prices.drop_duplicates()
```

```python
len(prices)
```

```python
prices = prices.reset_index()
prices = prices[['ticker','date','open','high','low','close','volume','dividend','split']]
```

```python
prices = prices.set_index('ticker')
```

```python
prices = prices.sort_values(['ticker','date'])
```

```python
prices
```

```python
for ticker in tickers:
    ticker_frame = prices[prices.index == ticker]
    ticker_frame.to_csv('/home/ubuntu/csvs/{}.csv'.format(ticker),index=False)
```
